```r
survey = read.csv(file="Day1Survey.csv", header=TRUE)
head(survey)
hist(x = survey$TV, main="Histogram of number of hours watching TV in a typical week",
xlab = "Number of hours watching TV in a typical week", 
     ylab = "Number of People", breaks = 25)
```

Summary The shape of this histogram is left skewed. The centre of this histogram is [3,4). The variability of this histogram is 0-10. There are two outlier bins in this histogram: [19,20) and [24,25)

```r
hist(x = survey$Pulse, main="Histogram of Resting Pulse Rate", 
     xlab = "Resting Pulse Rate", ylab = "Number of People")
```

Summary The shape of this histogram is bimodal. The centre of this histogram is [65,70). The variability of this histogram is 45-100. There are are no outliers in this histogram.
