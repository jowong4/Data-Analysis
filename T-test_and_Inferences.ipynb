```r
pollution = read.csv(file="AutoPollution_two_sizes.csv", header=TRUE)
head(pollution)
```

```r
small.pollution = subset(pollution, Size=="1")
hist(small.pollution$Noise, main="Distribution of Noise Levels", sub = "Size = 1",
     col="seagreen3", xlab = "Noise Level", ylab = "Number of Cars")
large.pollution = subset(pollution, Size=="3")
hist(large.pollution$Noise, main="Distribution of Noise Levels", sub = "Size = 3",
     col="seagreen3", xlab = "Noise Level", ylab = "Number of Cars")
```

```r
tapply(pollution$Noise, pollution$Size, sd)
```

```r
7.64/10.34
```

The ratio of the sample standard deviations is 0.74 which is between 0.5 and 2, and so we use the pooled version of the t-test (default)

Hypothesis:

H-zero : miu-large - miu-small = 0

H-one : miu-large - miu-small =/= 0

alpha-level = 0.05

```r
t.test(x=large.pollution$Noise, y=small.pollution$Noise, var.equal=TRUE, 
       conf.level = 0.95)
```

With a p-value of 2.17e-12, there is strong evidence for rejecting the null hypothesis. Therefore, I conclude that large and small cars have different noise level. Furthermore, because this is an experiment where cars are selected randomly, we can also conclude that the size is the cause of the difference in noise level.

The p-value of in the above means that with a null hypothesis of no difference in noise level between small and large cars, there is a 2.17e-12 probability that the difference in noise level between the small and large cars of -51.7 occur purely based on chance.

**D.**

```r
t.test(x=large.pollution$Noise, y=small.pollution$Noise,
       conf.level=0.90)
```

We have -51.7 (calculated above) as the point estimate for the difference in noise level between small and large cars. With a 90% confidence, the confidence interval for difference in noise level between small and large cars is [-58.06,-45.3]

We can conclude that with 90% confidence, the true noise level difference between small and large cars fall somewhere in [-58.06,-45.3]. And since the null hypothesis of no difference (0) lies outside the confidence interval, we also conclude that the noise level between small and large cars does not equal to 0.
